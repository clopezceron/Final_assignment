# Introduction

This project tries to create a program aimed to analyze the language in literary works by applying statistical methods to compare different texts. The main objective is to identify similarities and differences between the writing styles of the works through quantitative linguistic features. 

# General description of the program

The program was designed to analyze literary texts by extracting statistical information about their vocabulary and comparing the linguistic characteristics of different works. The general features of the program consists of the following:

## Text preprocessing

The function process_book() receives a string and returns four different representations of the book: the list of lemmas, the list of original words, the lemmatized text as a string, and the cleaned original text.

For example:

```python
gatos=process_book("Los gatos corrían rápidamente, y los gatos jugaban.")
gatos
```

## Frequency calculation

The function get_frecuencias() receives the list of original words and the list of lemmas and counts how many times each one appears.

Example:

```python
numero_gatos_y_perros=get_frecuencias( ["gato", "perro", "gato"], ["gato", "perro", "gato"])
numero_gatos_y_perros
```

## Similarity criterion

The function compare() evaluates the similarity between two literary works. It combines several indicators:

Similarity between the frequencies of the most common words.

Similarity between the frequencies of the most common lemmas.

Similarity between the distribution of characters.

Similarity in lexical diversity (ratio of different words).

Similarity in the proportion of unique words.

Similarity in the proportion of lemmas.

Similarity in the number of uncommon characters.

The comparison produces a score from 0 to 100, where a value closer to 100 indicates a higher linguistic similarity according to the proposed statistical criterion.


```python
files = ["hojarasca.txt", "coronel.txt", "cronicas.txt", "ruiseñor.txt"]

with open("dictionary.txt", "r", encoding="utf-8") as file:
    dictionary = file.read().lower().split()
```

```python
books = []
frequencies = []
lemmas_frequencies = []

for file in files:
    with open(file, "r", encoding="utf-8") as f:
            text = f.read().lower()
                
                    clean_book = process_book(text)

                        books.append(clean_book[3])

                            freq, freq_lemmas = get_frecuencias(
                                        clean_book[1],
                                                clean_book[0]
                            )

                                frequencies.append(freq)
                                    lemmas_frequencies.append(freq_lemmas)
```

The following outcomes presents the similarity scores obtained by comparing each pair of literary works using the proposed comparison criterion.

```python
import numpy as np

n = len(files)

comparison_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
            if i == j:
                        comparison_matrix[i, j] = 100
                                else:
                                            comparison_matrix[i, j] = compare(frequencies[i],lemmas_frequencies[i], frequencies[j], lemmas_frequencies[j], dictionary, books[i], books[j])
comparison_matrix
                                            
```

import pandas as pd
